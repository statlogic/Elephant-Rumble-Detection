```python
import numpy as np
from kafka import KafkaConsumer
```

```python
topic = 'test'
bootstrapServer = ['localhost:9092']
consumer = KafkaConsumer(topic,bootstrap_servers = bootstrapServer)
```

```python
AUDIO_LENGTH = 4
SAMPLE_RATE = 22050
TOTAL_SAMPLES_IN_GIVEN_TIME = AUDIO_LENGTH * SAMPLE_RATE

dataList = None

for message in consumer:
    data = message.value
    values = np.frombuffer(data, np.float32)
    
    if dataList is None:
        dataList = values
    else:
        dataList = np.append(dataList,values)
    
    if len(dataList) >= TOTAL_SAMPLES_IN_GIVEN_TIME:
        audioPatch = dataList[-TOTAL_SAMPLES_IN_GIVEN_TIME:]
        
```
